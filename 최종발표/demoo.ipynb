# 랜덤포레스트를 통한 훈련 및 학습

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import sklearn.metrics as metrics 

# 데이터를 Train, Test 데이터로 분할
X_train, X_test, Y_train, Y_test = train_test_split(pWeather5.iloc[:,3:-1],
                                                    pWeather5.iloc[:,-1:],
                                                    test_size = 0.2)
print(X_train)
```

```python
print(X_test)
```

# F1 Score를 통한 학습모델 성능 평가

```python
# 결과 반환 리스트
result = []


 # 랜덤포레스트 객체 생성
rfc = RandomForestClassifier(n_estimators=150,
                             min_samples_leaf=10,
                             n_jobs= -1,
                             max_features=12)
    
# 훈련
rfc.fit(X_train, Y_train.values.ravel())
# train과 test에 해당하는 예측결과

testPred = rfc.predict(X_test)        
result.append([metrics.f1_score(Y_test, testPred)])


print(result)
```
